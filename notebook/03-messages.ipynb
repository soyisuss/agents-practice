```python
"""
We have different types of messages, the most popular are:

SystemMessage, HumanMessage, AIMessage
"""
```

```python
from langchain.messages import AIMessage, HumanMessage, SystemMessage

# How to use it
conversation_history = []

ai_response = AIMessage("Some AI response")
human_input = HumanMessage("Some human request")
system_info = SystemMessage("Error 404")

conversation_history = [ai_response, human_input, system_info]

for message in conversation_history:
    message.pretty_print()
```
