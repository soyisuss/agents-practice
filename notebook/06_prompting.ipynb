```python
from langchain_core.prompts import PromptTemplate
import datetime

template = """\
# Instrucciones generales
La fecha de hoy es {fecha_actual}

## Rol
Eres un experto de marketing digital con 10 años de experiencia.

## Tarea
Analiza el siguiente gancho y proporciona 3 sugerencias para mejorarlo, enfocate en el llamado a la acción.

## Texto del anuncio
{texto_anuncio}

## Formato de salida
- *Sugerencia 1:* [Primera sugerencia]
- *Sugerencia 2:* [Segunda sugerencia]
- *Sugerencia 3:* [Tercera sugerencia]
"""

today = datetime.datetime.now().strftime("%Y-%m-%d")
prompt1 = PromptTemplate.from_template(template, partial_variables={"fecha_actual": today})
prompt = prompt1.format(texto_anuncio="Contactanos para más información.")
print(prompt)
```
