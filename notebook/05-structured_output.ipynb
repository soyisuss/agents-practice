```python
#from langchain.chat_models import init_chat_model
from langchain_deepseek import ChatDeepSeek

#model = init_chat_model(model="google_genai:gemini-2.5-flash-lite")
model = ChatDeepSeek(model="deepseek-chat")

response = model.invoke("Hola, Soy Yael y tengo 22 años")
response.pretty_print()
```

```python
from pydantic import BaseModel, Field

class PersonalInfo(BaseModel):
    """Personal user information"""
    name: str = Field(description="The name of the person")
    age: int = Field(description="The aghe of the person")

messages = [
    ("system", "Eres un asistente que se encarga de extraer la informacion personal del usuario de una conversacion"),
    ("user", "Hola, Soy Yael y tengo 22 años")
]

model_structured = model.with_structured_output(schema=PersonalInfo)
response = model_structured.invoke(messages)
response
```
